### Filling Null Values

My overall strategy for filling the null values was to be cognizant of how filling a null value could impact the data set. For example, the variable alcohol is described in the data set as: '*recorded per capita(15+) consumption (in litres of pure alcohol)*. I didn't want to fill that variable with the mean because that value seemed ***high*** for some countries missing that value. So instead, I chose the median. 

For each of the variables in empty_list, if it made sense to do so I used the more **conservative value** between mean and median, in an effort to minimize the impact of filling the data.

**Alcohol:** A lot of countries in the world have very low **official and unofficial** levels of alcohol consumption. I wanted to attempt to preserve that perspective, so I chose the mean.

**Hepatitis B:** Geographically, I'm not sure on how the likelihood of contracting Hepatitis B varies by location. If we're comparing two places that have drastically different likelihoods of contraction and the place with a lower likelihood of contraction is null, I don't want to add the higher number.

**Total Expenditure:** The difference seemed negligible so I used the mean.

**GDP:** The mean seemed high to me, so I chose the more conservative median.

**Population:** My logic here was in most places, people are probably under counted. In this case the median was higher so I chose that number.

**Income Composition of Resources:** The difference seemed negligible so I used the mean.

**Schooling:** My logic here was that there are probably fewer people in school than are actually counted as being in school. So I chose the lower number.

***Handling Outliers***

I used the **Tukey's Method** to identify outliers in all of the numeric columns except for year. Any values above or below the top and bottom range of our interquartile range are considered outliers.

```python
q75, q25 = np.percentile(life_df.iloc[:,3:20], [75 ,25])
iqr = q75 - q25

for column_name in life_df.iloc[:,3:20]:
    print("-----------------------------------------")
    print("In {}:".format(column_name))
    print("-----------------------------------------")
    for threshold in np.arange(1,5,0.5):
        min_val = q25 - (iqr*threshold)
        max_val = q75 + (iqr*threshold)
        print("The score threshold is: {}".format(threshold))
        print("Number of outliers is: {}".format(
        len((np.where((life_df[column_name] > max_val) 
                      | (life_df[column_name] < min_val))[0]))
    ))
```

**How Do the Statuses Interact with Life Expectancy**
Even though the mean's of Develop and Developing are statistically significant. There ***is*** also a statistically significant difference regarding how those variables and life expectancy.

```python
status = life_df["status"].unique()
grouped_df = life_df.groupby("status")

for var in ["life expectancy"]:
    print("------------------------------------------------")
    print("Comparisons for variable: {}".format(var))
    print("------------------------------------------------")
    for i in range(0, len(status)):
        for j in range(i+1, len(status)):
            print("t-test between groups {0} and {1}:".format(status[i], status[j]))
            print(stats.ttest_ind(
                life_df[life_df["status"]== status[i]][var], 
                life_df[life_df["status"]== status[j]][var]
            ))
```
