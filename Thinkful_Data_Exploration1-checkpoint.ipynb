```python
fill_list = ['AVG_MATH_4_SCORE','AVG_MATH_8_SCORE','AVG_READING_4_SCORE','AVG_READING_8_SCORE']
print(fill_list)
```

```python
co2_df = co_df.copy()
for col in fill_list:
    co2_df.loc[:, col].interpolate(inplace = True)
```

```python
co3_df = co_df.copy()
for col in fill_list:
    co3_df.loc[:, col].fillna(co3_df.loc[:,col].mean(),inplace = True)
```

```python
co4_df = co_df.copy()
for col in fill_list:
    co4_df.loc[:, col].fillna(co4_df.loc[:,col].median(),inplace = True)
```

```python
for col in fill_list:
    print("Statistics for columns: {}".format(col))
    print(pd.concat([co_df[col],co2_df[col],co3_df[col],co4_df[col]], axis = 1).describe())
```

**Notice there are too many missing values for math and reading scores. Fill out the missing values using mean, median, and linear interpolation. Then compare the effects of these techniques on the distributions of the score variables.**
It doesn't seem that filling the null values with the methods above changed the score distributions in Colorado.

```python
fig = plt.figure(figsize=(15,20))

for i, col in enumerate(fill_list):
    df = pd.concat([co_df[col], co2_df[col],
                     co3_df[col], co4_df[col]], axis=1)
    
    plt.subplot(len(fill_list), 4, i+1)
    plt.bar(["original", "interpolation", "mean", "median"], df.median())
    plt.xticks(rotation='vertical')
    plt.title("Mod of {}".format(col))
    
plt.tight_layout()
plt.show()
```
