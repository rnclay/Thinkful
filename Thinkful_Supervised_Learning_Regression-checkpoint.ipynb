### Cleaning the Data

Let's get a sense for how many null values are in this data set.

```python
houseprice_df.isna().sum()*100/houseprice_df.isna().count()

```

Let's take a closer look at the unique values in each of the columns that contain null values.

```python
empty_values = ["lotfrontage","alley","masvnrtype","masvnrarea","fireplacequ","garagetype",
               "garageyrblt","garagefinish","garagequal","garagecond","poolqc","fence","miscfeature"]
for column_name in empty_values:
    print("Unique values in column {} are: {}".format(column_name, houseprice_df[column_name].unique()))
```

It looks like most ofthe columns with null values are **categorical**. So now we should group these variables into numerical and categorical and fill the values of each group.

But first let's check the numerical columns for any non-float object types

```python
empty_floats = ["lotfrontage","masvnrarea","garageyrblt"]

for column_name in empty_floats:
    print("-----These are the problematic values for the variable----:{}".format(column_name))
    for value in houseprice_df[column_name]:
        try:
            float(value)
        except:
            print(value)
```

Let's fill in the null values for our continuous and time series variables.

```python
houseprice_df["lotfrontage"].fillna(houseprice_df["lotfrontage"].mean(), inplace=True)
houseprice_df["masvnrarea"].fillna(houseprice_df["masvnrarea"].mean(), inplace=True)
houseprice_df["garageyrblt"].interpolate(method='linear',inplace=True)
```

Now we can address the null values of our categorical variables

```python
houseprice_df['alley'].fillna("no_alley", inplace=True)
houseprice_df['masvnrtype'].fillna("no_masonry", inplace=True)
houseprice_df['fireplacequ'].fillna("no_fireplace", inplace=True)
houseprice_df['garagetype'].fillna("no_garage", inplace=True)
houseprice_df['garagefinish'].fillna("no_garage", inplace=True)
houseprice_df['garagequal'].fillna("no_garage", inplace=True)
houseprice_df['garagecond'].fillna("no_garage", inplace=True)
houseprice_df['poolqc'].fillna("no_pool", inplace=True)
houseprice_df['miscfeature'].fillna("no_misc", inplace=True)
houseprice_df['fence'].fillna("no_fence", inplace=True)
```

### Exploring the Data

We need to check for outliers, so let's break our continous variables out of our data frame.

```python
continuous=houseprice_df.select_dtypes(np.number).head()
```

```python
q75, q25 = np.percentile(continuous, [75 ,25])
iqr = q75 - q25

for column_name in continuous:
    print("-----------------------------------------")
    print("In {}:".format(column_name))
    print("-----------------------------------------")
    for threshold in np.arange(1,5,0.5):
        min_val = q25 - (iqr*threshold)
        max_val = q75 + (iqr*threshold)
        print("The score threshold is: {}".format(threshold))
        print("Number of outliers is: {}".format(
        len((np.where((continuous[column_name] > max_val) 
                      | (continuous[column_name] < min_val))[0]))
    ))
```

The only variable with outliers in our continuous variable group is 'sales price', with a total of 5 outliers.

I've found narrowing down the categorical variables more troublesome than narrowing down my continuous variables. At this point, I plan to evaluate the unique values in each column, if a column has ***one*** unique value, it will ***NOT*** be selected. 

If there is only one unique value, then it is common to all and there is no instances of that variable being absent, there is nothing to compare it against.

Variables with ***three*** or ***more*** values will be evaluated on a per case basis. If they cannot be converted into a binary relationship, then they will be dropped.

Variables with two unique values will be evaluated.
